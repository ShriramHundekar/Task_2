# Titanic Dataset (Cleaned) - Exploratory Data Analysis (EDA)
This notebook performs EDA on the pre-cleaned Titanic dataset.

### Step 1: Load the Cleaned Dataset

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load cleaned dataset
df = pd.read_csv("cleaned_titanic_data.csv")
df.head()
```

### Step 2: Summary Statistics

```python
df.describe()
```

### Step 3: Histograms and Boxplots

```python
# Age distribution
sns.histplot(df['Age'], bins=30, kde=True)
plt.title("Distribution of Age (Standardized)")
plt.show()

# Fare distribution
sns.histplot(df['Fare'], bins=30, kde=True)
plt.title("Distribution of Fare (Standardized)")
plt.show()

# Boxplot for Fare
sns.boxplot(x=df['Fare'])
plt.title("Boxplot of Fare")
plt.show()
```

### Step 4: Correlation Heatmap

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

### Step 5: Survival by Gender and Class

```python
sns.barplot(x='Sex_male', y='Survived', data=df)
plt.title("Survival Rate by Gender (0=Female, 1=Male)")
plt.show()

sns.barplot(x='Pclass', y='Survived', data=df)
plt.title("Survival Rate by Passenger Class")
plt.show()
```

### Step 6: Observations & Inferences
- Females (Sex_male = False) had higher survival.
- 1st class passengers had better survival odds.
- Some correlation seen between Fare and Survived.
